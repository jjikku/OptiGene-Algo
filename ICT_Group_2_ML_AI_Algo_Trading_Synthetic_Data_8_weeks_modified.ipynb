```python
import pandas as pd
import ta

df_TA = df_synthetic.copy()
df_TA['Date/Time'] = pd.to_datetime(df_TA['Date/Time'])
df_TA.set_index('Date/Time', inplace=True)

# Define the windows for each timeframe in minutes
windows = {
    '3m': 3,
    '5m': 5,
    '7m': 7,
    '10m': 10,
    '15m': 15,
}

# Function to calculate indicators for each group
def calculate_indicators(group):
    # Make sure the index is carried through the operations
    group = group.copy()
    # Calculate the indicators for each window
    for label, window in windows.items():
        # Calculate the MACD with a span of the given window
        macd = ta.trend.MACD(group['Close'], window_slow=window, window_fast=window//2, window_sign=window//2)
        group[f'MACD_{label}'] = macd.macd()

        # Calculate the RSI
        group[f'RSI_{label}'] = ta.momentum.RSIIndicator(group['Close'], window=window).rsi()

        # Calculate the EMA
        group[f'EMA_{label}'] = ta.trend.EMAIndicator(group['Close'], window=window).ema_indicator()

        # Calculate the Stochastic Oscillator
        stoch = ta.momentum.StochasticOscillator(group['Close'], group['Close'], group['Close'], window=window, smooth_window=window//2)
        group[f'Stoch_%K_{label}'] = stoch.stoch()
        group[f'Stoch_%D_{label}'] = stoch.stoch_signal()

        # Calculate the ATR
        group[f'ATR_{label}'] = ta.volatility.AverageTrueRange(group['High'], group['Low'], group['Close'], window=window).average_true_range()

         # Bollinger Bands
        bollinger = ta.volatility.BollingerBands(group['Close'], window=window, window_dev=2)
        group[f'Bollinger_High_{label}'], group[f'Bollinger_Mid_{label}'], group[f'Bollinger_Low_{label}'] = bollinger.bollinger_hband(), bollinger.bollinger_mavg(), bollinger.bollinger_lband()

        # MACD Histogram
        group[f'MACD_Hist_{label}'] = macd.macd_diff()

    return group

# Group by each day, but keep the original index
df_grouped = df_TA.groupby(df_TA.index.date).apply(calculate_indicators)

# Check for NaN values after calculations
nan_count = df_grouped.isna().any(axis=1).sum()
print("drop count", nan_count)

# Drop rows with NaN values
# df_grouped.dropna(inplace=True)

# The DataFrame index is the 'Date/Time' column
# If you want 'Date/Time' as a regular column instead of the index, you can reset the index
df_grouped.reset_index(inplace=True)

# Display the head of the dataframe
df_grouped.head(50)
```

```python
import pandas as pd
import numpy as np

def flag_entries_with_confidence(df, threshold=30):
    if df['Date/Time'].dtype == 'object':
        df['Date/Time'] = pd.to_datetime(df['Date/Time'])

    df.sort_values('Date/Time', inplace=True)
    df['CE_Buys'] = 0
    df['PE_Buys'] = 0
    df['CE_Confidence_Score'] = 0.0
    df['PE_Confidence_Score'] = 0.0

    import numpy as np

    def calculate_confidence_score(current_close, future_closes, threshold=30, scale_max=1, scale_min=-1):
        def linear_scale(move, threshold, scale_max, scale_min):
            if abs(move) <= threshold:
                return move / threshold * scale_max
            else:
                return scale_max if move > 0 else scale_min

        # Initialize scores
        ce_score = 0.0
        pe_score = 0.0

        # Track the previous move to detect change in direction
        prev_move = 0

        # Assess trend over each of the next 5 candles
        for i, future_close in enumerate(future_closes):
            move = future_close - current_close
            scaled_move = linear_scale(move, threshold, scale_max, scale_min)
            ce_bias = 1.5
            pe_bias = 1.5

            # Check for change in direction
            if move > 0:
                ce_score += scaled_move * ce_bias * (1 - i / 5)
                if prev_move <= 0:
                    pe_score -= scaled_move * (i / 5)
            else:
                pe_score += abs(scaled_move) * pe_bias * (1 - i / 5)
                if prev_move >= 0:
                    ce_score -= abs(scaled_move) * (i / 5)
            prev_move = move

        return ce_score, pe_score



    def apply_checks(group):
        last_ce_buy_index = -5
        last_pe_buy_index = -5

        for i in range(len(group) - 5):
            future_closes = group['Close'].iloc[i+1:i+6]
            slice_ = group.iloc[i:i+5]
            ce_score, pe_score = calculate_confidence_score(group['Close'].iloc[i], future_closes)
            group.at[group.index[i], 'CE_Confidence_Score'] = ce_score
            group.at[group.index[i], 'PE_Confidence_Score'] = pe_score

            if i - last_ce_buy_index >= 5 or i - last_pe_buy_index >= 5:
                if all(slice_['Close'].diff().iloc[1:] > 0) and (slice_['Close'].iloc[-1] - slice_['Close'].iloc[0] > threshold):
                    group.at[slice_.index[0], 'CE_Buys'] = 1 if ce_score > 0 else 0
                    last_ce_buy_index = i
                elif all(slice_['Close'].diff().iloc[1:] < 0) and (slice_['Close'].iloc[0] - slice_['Close'].iloc[-1] > threshold):
                    group.at[slice_.index[0], 'PE_Buys'] = 1 if pe_score > 0 else 0
                    last_pe_buy_index = i

        return group

    # Apply the checks to each group and concatenate
    df = pd.concat([apply_checks(group) for _, group in df.groupby(df['Date/Time'].dt.date)])

    return df

df_synthetic = flag_entries_with_confidence(df_grouped)
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Function to create and customize a candlestick chart with additional indicators
def create_tradingview_like_chart(df):
    fig_height = 600  # Height of the figure in pixels

    # Create a subplot figure with 3 rows
    fig = make_subplots(rows=3, cols=1, shared_xaxes=True,
                        vertical_spacing=0.02,
                        subplot_titles=('Candlestick', 'PCR', 'PCR Volume'),
                        row_heights=[0.6, 0.2, 0.2])

    # Add the candlestick chart to the first row
    fig.add_trace(go.Candlestick(
        x=df['Date/Time'],
        open=df['Open'],
        high=df['High'],
        low=df['Low'],
        close=df['Close'],
        increasing_line_color='green',
        decreasing_line_color='red',
        increasing_fillcolor='green',
        decreasing_fillcolor='red'
    ), row=1, col=1)

    # Add PE Buys entry points
    fig.add_trace(go.Scatter(
        x=df['Date/Time'][df['PE_Buys_conf'] == 1],
        y=df['High'][df['PE_Buys_conf'] == 1] + 2,
        mode='markers',
        name='PE Buys',
        marker=dict(symbol='triangle-down', size=10, color='rgba(128, 0, 128, 1)', line=dict(width=2, color='rgba(153, 50, 204, 1)'))
    ), row=1, col=1)

    # Add CE Buys entry points
    fig.add_trace(go.Scatter(
        x=df['Date/Time'][df['CE_Buys_conf'] == 1],
        y=df['Low'][df['CE_Buys_conf'] == 1] - 2,
        mode='markers',
        name='CE Buys',
        marker=dict(symbol='triangle-up', size=10, color='rgba(0, 0, 255, 1)', line=dict(width=2, color='rgba(65, 105, 225, 1)'))
    ), row=1, col=1)

    # Add PCR plot to the second row
    fig.add_trace(go.Scatter(
        x=df['Date/Time'],
        y=df['PCR'],
        line=dict(color='blue', width=2),
        name='PCR'
    ), row=2, col=1)

    # Add PCR Volume plot to the third row
    fig.add_trace(go.Scatter(
        x=df['Date/Time'],
        y=df['PCR_Volume'],
        line=dict(color='orange', width=2),
        name='PCR Volume'
    ), row=3, col=1)

    # Customize layout for a more TradingView-like style
    fig.update_layout(
        paper_bgcolor='rgba(18, 18, 18, 1)',
        plot_bgcolor='rgba(18, 18, 18, 1)',
        xaxis_rangeslider_visible=False,
        height=fig_height,
        font=dict(family="Verdana, sans-serif", size=12, color="white")
    )

    # Customize x-axis and y-axis
    fig.update_xaxes(gridcolor='rgba(68, 68, 68, 0.5)', showline=True, mirror=True, linecolor='rgba(68, 68, 68, 1)', rangeselector=dict(
        buttons=list([
            {'count': 3, 'label': '3m', 'step': 'minute', 'stepmode': 'backward'},
            {'count': 5, 'label': '5m', 'step': 'minute', 'stepmode': 'backward'},
            {'count': 10, 'label': '10m', 'step': 'minute', 'stepmode': 'backward'},
            {'count': 15, 'label': '15m', 'step': 'minute', 'stepmode': 'backward'},
            {'count': 1, 'label': '1M', 'step': 'month', 'stepmode': 'backward'},
            {'count': 6, 'label': '6M', 'step': 'month', 'stepmode': 'backward'},
            {'step': 'all'}
        ]),
        bgcolor="rgba(0,0,0,0.1)"
    ))

    # Show the figure
    fig.show()

create_tradingview_like_chart(df_OHLC)
```
